```python
# NYC Citibike tripdata History Logs from February 2020 - April 2020
import pandas as pd
import datetime
import numpy as np
import os
```

```python
# Load NYC Citibike tripdata csv files from 202002 to 202004
trip_feb = os.path.join("resources" ,"JC-202002-citibike-tripdata.csv")
trip_mar = os.path.join("resources" ,"JC-202003-citibike-tripdata.csv")
trip_apr = os.path.join("resources" ,"JC-202004-citibike-tripdata.csv")

nycbike_feb = pd.read_csv(trip_feb)
nycbike_mar = pd.read_csv(trip_mar)
nycbike_apr = pd.read_csv(trip_apr)
```

```python
nycbike_feb.head(10)
```

```python
nycbike_feb_mar_apr = pd.concat([nycbike_feb,nycbike_mar,nycbike_apr])
```

```python
# Output concatenated 2018 July-Aug citibike tripdata to CSV file 
filepath_2020 = os.path.join("Resources" ,"JC-2020.csv")
nycbike_feb_mar_apr.to_csv(filepath_2020, index=False)
nycbike_feb_mar_apr.tail(10)
```

```python
nycbike_feb_mar_apr.rename(columns = {'Trip Duration': 'tripduration', 'Start Time': 'starttime', 
                      'Stop Time': 'stoptime', 'Gender':'gender', "User Type": "usertype",
                      'Bike ID': 'bikeid', 'Birth Year': 'birth year', 'Start Station ID': 'start station id',
                      'Start Station Name': 'start station name', 'Start Station Latitude': 'start station latitude',
                      'Start Station Longitude': 'start station longitude', 'End Station ID': 'end station id', 
                      'End Station Name': 'end station name', 'End Station Latitude': 'end station latitude',
                      'End Station Longitude': 'end station longitude'}, inplace = True)

nycbike_feb_mar_apr.head()
```
